```python
import pandas as pd
import json
import os
```

```python
path2 = "D:\\phonepe_project\\pulse\\data\\aggregated\\user\\country\\india\\state"
path_a = os.listdir(path2)
```

```python
std2 = {'state':[], 'year':[],'quater':[],'brand_name':[], 'brand_count':[], 'brand_usage':[]}


for i in path_a:
    path_yr = path2 +"\\"+i+"\\"
    year = os.listdir(path_yr)

    for j in year:
        path_file = path_yr + j
        path_n = os.listdir(path_file)

        for k in path_n:
            asd = path_file + "\\" + k
            data_n = open(asd, "r")
            output = json.load(data_n)

            for l in output['data']:
                name = l['usersByDevice'][0]['brand']
                count = l['usersByDevice'][0]['count']
                usage = l['usersByDevice'][0]['percentage']
                std2['brand_name'].append(name)
                std2['brand_count'].append(count)
                std2['brand_usage'].append(usage)
                std2['state'].append(i)
                std2['year'].append(j)
                std2['quater'].append(int(k.strip('.json')))

            
df2 = pd.DataFrame(std2)
```

```python
std2 = {'state': [], 'year': [], 'quarter': [], 'brand_name': [], 'brand_count': [], 'brand_usage': []}

for i in path_a:
    path_yr = path2 + "\\" + i + "\\"
    year = os.listdir(path_yr)

    for j in year:
        path_file = path_yr + j
        path_n = os.listdir(path_file)

        for k in path_n:
            asd = path_file + "\\" + k
            with open(asd, "r") as data_n:
                try:
                    output = json.load(data_n)
                except json.JSONDecodeError as e:
                    print(f"Error loading JSON in {asd}: {e}")
                    continue

            if output and output.get('data') and output['data'].get('usersByDevice'):
                for l in output['data']['usersByDevice']:
                    name = l.get('brand', 'Unknown Brand')
                    count = l.get('count', 0)
                    usage = l.get('percentage', 0.0)
                    std2['brand_name'].append(name)
                    std2['brand_count'].append(count)
                    std2['brand_usage'].append(usage)
                    std2['state'].append(i)
                    std2['year'].append(j)
                    std2['quarter'].append(int(k.strip('.json')))
            else:
                print(f"Missing data in {asd}")

df2 = pd.DataFrame(std2)
```

```python
import pandas as pd
from sqlalchemy import create_engine

# PostgreSQL database connection parameters
db_config = {
    'user': 'postgres',
    'password': '123456',
    'host': 'localhost',
    'port': '5432',
    'database': 'postgres',
}

# Create a SQLAlchemy engine
engine = create_engine(f'postgresql+psycopg2://{db_config["user"]}:{db_config["password"]}@{db_config["host"]}:{db_config["port"]}/{db_config["database"]}')

# CSV file path
csv_file_path = 'D:\\phonepe_project\\agg_user.csv'

# Read CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# Insert data into PostgreSQL table
try:
    df.to_sql('agg_user', con=engine, if_exists='replace', index=True)
    print("Data inserted successfully.")
except Exception as e:
    print("Error:", e)
finally:
    engine.dispose()  # Close the database connection
```
